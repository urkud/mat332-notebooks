```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
Y, X = np.mgrid[-7:7:15j, -7:7:20j]
dX = np.ones(X.shape)
dY = 1 - Y
len_dXY = np.sqrt (dX ** 2 + dY ** 2)
dXn = dX / len_dXY
dYn = dY / len_dXY

plt.quiver(X, Y, dXn, dYn, headlength=0, headwidth=1, pivot='middle')
plt.scatter(X, Y, 15, fc='white', ec='black')
```
